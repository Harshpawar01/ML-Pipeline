```python
import pandas as pd

data = pd.read_csv('/content/heart_disease_uci.csv')

print(data.head())

print(data.info())
print(data.isnull().sum())
```

question no 2

```python
data['thal'].fillna(data['thal'].mode()[0], inplace=True)
data['ca'].fillna(data['ca'].mode()[0], inplace=True)

print(data.isnull().sum())
```

```python
data.dropna(inplace=True)
```

question no 3

```python
data['AgeGroup'] = pd.cut(data['age'], bins=[0, 40, 60, 120], labels=['<40', '40-60', '>60'])
```

```python
data['CholesterolLevel'] = pd.cut(data['chol'], bins=[0, 200, 239, 600], labels=['Low', 'Normal', 'High'])
```

```python
data['IsRisk'] = ((data['age'] > 60) & (data['chol'] > 239) & (data['trestbps'] > 140)).astype(int)
```

question no 4

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

le = LabelEncoder()
data['sex'] = le.fit_transform(data['sex'])
data['thal'] = le.fit_transform(data['thal'])

data = pd.get_dummies(data, columns=['cp', 'AgeGroup'])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data[['chol', 'trestbps', 'thalch']] = scaler.fit_transform(data[['chol', 'trestbps', 'thalch']])
```
